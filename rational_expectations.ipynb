# A Rational Expectations Model

This notebook follows Ljungqvist and Sargent's Rational expectations. 

```python
import numpy as np
from quantecon import LQ
import matplotlib.pyplot as plt
%matplotlib inline
```

# Planner's Problem
\begin{equation}
x_t = \begin{bmatrix} Y_t \\ 1 \end{bmatrix},
\quad
u_t = Y_{t+1} - Y_{t} \\
A = \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix},
\quad
B = \begin{bmatrix} 1 \\ 0 \end{bmatrix},
\quad
R = \begin{bmatrix} a_1/2 & -a_0/2 \\ -a_0/2 & 0 \end{bmatrix},
\quad
Q = \gamma / 2
\end{equation}

```python
# == parameters == #
a0=100
a1=0.05
beta=0.95
gamma=10

# == setting up lq structure == #
R = np.array([[a1/2, -a0/2],
            [-a0/2, 0]])

Q = gamma/2

A = np.identity(2)

B = np.array([[1],
            [0]])

# == solving == #

mdl = LQ(Q, R, A, B, beta = beta)
x0 = (1, 1)
xs, us, ws = mdl.compute_sequence(x0)
P, F, d = mdl.stationary_values()

# == plotting == #
fig, ax = plt.subplots(figsize = (9, 6))
ax.plot(xs[0, ])
ax.set_title("Output Choices by a Firm (As per a Social Planner)")
plt.show()
```

```python
F = F.flatten()
k0, k1 = -F[1], 1 - F[0]
print(k0, k1)
```

The planner's problem and the market problem give the same answer. We just verified that these values of k0 and k1 are the parameters which give us the RE equilibrium.

When solving the market problem, we just guessed these parameters and got out a law of motion for Y. But solving the planner's problem gives us the RE equilibrium compatible parameter values.

# Defining the market problem as a LQDP

\begin{equation}
x_t = \begin{bmatrix} y_t \\ Y_t \\ 1 \end{bmatrix},
\qquad
u_t = y_{t+1} - y_{t}
\\
A =
\begin{bmatrix}
    1 & 0 & 0 \\
    0 & \kappa_1 & \kappa_0 \\
    0 & 0 & 1
\end{bmatrix},
\quad
B = \begin{bmatrix} 1 \\ 0 \\ 0 \end{bmatrix} ,
\quad
R =
\begin{bmatrix}
    0 & a_1/2 & -a_0/2 \\
    a_1/2 & 0 & 0 \\
    -a_0/2 & 0 & 0
\end{bmatrix},
\quad
Q = \gamma / 2
\end{equation}

```python
# == Params == #

a_0 = 100
a_1 = 0.05
beta = 0.95
gamma = 10

# == Beliefs == #
k_0 , k_1 = k0, k1

# == LQ Problem Matrices == #

A = np.array([[1, 0, 0],
            [0, k_1, k_0],
            [0, 0, 1]])

B = np.array([[1],
            [0],
            [0]])

R = np.array([[0, a_1/2, -a_0/2],
             [a_1/2, 0, 0],
             [-a_0/2, 0, 0]])

Q = gamma/2

# == Solving == #

mdl = LQ(Q, R, A, B, beta = beta)
x0 = (1, 1, 1)
xm, um, wm = mdl.compute_sequence(x0) #'m' in variable stands for 'market'
P, F, d = mdl.stationary_values() #this can be called without compute_sequence

# == plotting == #
fig, ax = plt.subplots(figsize = (9, 6))
ax.plot(xm[1, ])
ax.set_title("Output Choices by a Firm (As per the Market)")
plt.show()
```

```python
h0 = -F[0][2]
h1 = 1 - F[0][0]
h2 = -F[0][1]

print(f"""h0 = {h0} \nh1 = {h1} \nh2 = {h2} \n
The policy function looks like: \n
y_(t+1) = {h0} + {h1}*y_t + {h2}*Y_t""")
```

## Market vs Planner Output Paths

```python
# == plotting == #
fig, ax = plt.subplots(figsize = (9, 6))
ax.plot(xs[0, ], color='black', label = "Social Planner Output Path")
ax.plot(xm[1, ], marker='o', label = "Market Output Path", alpha=0.2)
ax.set_title("Output Choices")
ax.legend()
plt.show()
```

We used the parameters gained from the social planner problem in the market problem. Solving them yeilded the same result (as it should). If the agents in the model know as much as the social planner (the modeler i.e. we) do, then they take the same decisions as the planner.
